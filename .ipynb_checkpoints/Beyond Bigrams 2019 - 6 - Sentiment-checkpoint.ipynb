```python
# Standard imports
import sys
import os
import pandas as pd
import numpy as np
import textwrap
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
SEED=1

# Plotly
from plotly.offline import plot, iplot, init_notebook_mode
import plotly.graph_objs as go
import plotly.io as pio
init_notebook_mode(connected=True)

def matplotlib_to_plotly(cmap, pl_entries):
    h = 1.0/(pl_entries-1)
    pl_colorscale = []

    for k in range(pl_entries):
        C = list(map(np.uint8, np.array(cmap(k*h)[:3])*255))
        pl_colorscale.append([k*h, 'rgb'+str((C[0], C[1], C[2]))])

    return pl_colorscale

# Cell timing
# https://github.com/cpcloud/ipython-autotime
%load_ext autotime
```

```python
# Load data from Step 2
df = pd.read_csv('data/data-opiniosis-topics.txt', sep='\t')
vectors = np.load('data/vectors-opiniosis.npy')

# Filter to just the "hotel" cluster, to reduce the complexity
idx = (df.x < 0) & (df.topic_hotel > 0)
df_hotel = df.loc[idx]
vectors_hotel = vectors[idx,:]
```

# Textblob sentiment

```python
from textblob import TextBlob

s = []
for t in df_hotel.text:
    blob = TextBlob(t)
    sentiment = blob.sentiment.polarity
    s.append(sentiment)
    
df_hotel['sentiment_textblob'] = s
```

# OpenAI

```python
# Get the openAI sentiment
os.chdir("git/generating-reviews-discovering-sentiment/")
from encoder import Model
model = Model()
```

```python
results = model.transform(df_hotel["text"].tolist())
```

```python
df_hotel['sentiment_openai'] = results[:, 2388]
```

# NLTK Vadar

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
s = [sid.polarity_scores(t)['compound'] for t in df_hotel.text.tolist()]
df_hotel['sentiment_vadar'] = s
```

# Do something useful with it

```python
for label, group in df_hotel.groupby('topic_general'):
    print(label)
    print(group.sort_values(by='sentiment_openai', ascending=True).head(1).text.tolist())
    print(group.sort_values(by='sentiment_openai', ascending=False).head(1).text.tolist())
    print()
```

# Visualise

```python
plt.figure(figsize=(18,6))

plt.subplot(131)
plt.scatter(df_hotel.x, df_hotel.y, c=df_hotel.sentiment_textblob, s=5, vmin=-1, vmax=1, cmap='RdYlGn')
plt.title('Textblob')

plt.subplot(132)
plt.scatter(df_hotel.x, df_hotel.y, c=df_hotel.sentiment_vadar, s=5, vmin=-1, vmax=1, cmap='RdYlGn')
plt.title('Vadar')

plt.subplot(133)
plt.scatter(df_hotel.x, df_hotel.y, c=df_hotel.sentiment_openai, s=5, vmin=-1, vmax=1, cmap='RdYlGn')
plt.title('OpenAI')

plt.show()
```

```python
plt.figure(figsize=(18,6))

plt.subplot(131)
plt.scatter(df_hotel.x, df_hotel.y, c=df_hotel.sentiment_textblob<np.median(df_hotel.sentiment_textblob), s=5, vmin=0, vmax=1, cmap='RdYlGn')
plt.title('Textblob')

plt.subplot(132)
plt.scatter(df_hotel.x, df_hotel.y, c=df_hotel.sentiment_vadar<np.median(df_hotel.sentiment_vadar), s=5, vmin=0, vmax=1, cmap='RdYlGn')
plt.title('Vadar')

plt.subplot(133)
plt.scatter(df_hotel.x, df_hotel.y, c=df_hotel.sentiment_openai<np.median(df_hotel.sentiment_openai), s=5, vmin=0, vmax=1, cmap='RdYlGn')
plt.title('OpenAI')

plt.show()
```

```python
df_hotel['diff'] = (df_hotel.sentiment_openai - df_hotel.sentiment_textblob)
print('OpenAI low, Textblob high')
print(df_hotel.sort_values(by='diff').iloc[0][['text','sentiment_openai','sentiment_textblob']].values)
print('OpenAI high, Textblob low')
print(df_hotel.sort_values(by='diff', ascending=False).iloc[0][['text','sentiment_openai','sentiment_textblob']].values)
```

```python
# Plotly version for manual topic discovery
layout = go.Layout(
    autosize=False,
    width=800,
    height=800,
    hovermode="closest"
)

data = [go.Scatter(
                x=df_hotel.x,
                y=df_hotel.y,
                text=df_hotel.textwrap,
                mode='markers',
                marker=dict(
                        size=5,
                        colorscale=matplotlib_to_plotly(plt.cm.RdYlGn, 255),
                        color=df_hotel.sentiment_openai,
                        showscale=True
                )
)]

fig = go.Figure(data=data, layout=layout)

p = iplot(fig)
```

```python
# Plotly version for manual topic discovery
layout = go.Layout(
    autosize=False,
    width=800,
    height=800,
    hovermode="closest"
)

data = [go.Scatter(
                x=df_hotel.x,
                y=df_hotel.y,
                text=df_hotel.textwrap,
                mode='markers',
                marker=dict(
                        size=5,
                        colorscale=matplotlib_to_plotly(plt.cm.RdYlGn, 255),
                        color=(df_hotel.sentiment>0).astype(np.int),
                        showscale=True
                )
)]

fig = go.Figure(data=data, layout=layout)

p = iplot(fig)
```
