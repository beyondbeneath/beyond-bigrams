```python
# Load data from Step 2
df = pd.read_csv('output/data-opiniosis-topics.txt', sep='\t')
vectors = np.load('output/vectors-opiniosis.npy')

# Filter to just the "hotel" cluster, to reduce the complexity
idx = (df.x < 0) & (df.topic_hotel > 0)
df_hotel = df.loc[idx]
vectors_hotel = vectors[idx,:]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV, KFold

# Name things per convention
X = vectors_hotel
y = df_hotel.topic_hotel

# Create a KNN model
knn = KNeighborsClassifier(n_neighbors=3)

# Train
cv_scores = cross_val_score(knn, X, y, cv=5)
print(cv_scores)
print(cv_scores.mean())
```

```python
# Optimise the nneighbours
params = {'n_neighbors': np.arange(1, 25)}
knn = KNeighborsClassifier()
knn_search = GridSearchCV(knn, params, cv=5)

# Train
knn_search.fit(X, y)
```

```python
df_cv = pd.DataFrame(knn_search.cv_results_)

plt.figure(figsize=(12,6))
plt.plot(df_cv.param_n_neighbors, df_cv.mean_test_score, "k-", marker='o')
plt.xlabel('n_neighbors')
plt.ylabel('CV accuracy')
plt.show()
```

```python
# Create a KNN model
knn = KNeighborsClassifier(n_neighbors=15)

# Train
cv_scores = cross_val_score(knn, X, y, cv=5)
print(cv_scores)
print(cv_scores.mean())
```

```python
# Distance to nearest topic center
from sklearn.metrics.pairwise import cosine_similarity

kfold = KFold(n_splits=5, shuffle=True, random_state=0)
scores = []
for train_index, test_index in kfold.split(X):
    # Format datasets according to train/test
    X_train, y_train, df_hotel_train = X[train_index, :], y[train_index], df_hotel.iloc[train_index]
    X_test, y_test, df_hotel_test = X[test_index, :], y[test_index], df_hotel.iloc[test_index]
    
    # Find topic averages using train
    topic_averages = []
    for i in range(9):
        topic_id = i+1
        topic_averages.append(X_train[df_hotel_train.topic_hotel==topic_id, :].mean(axis=0))
    topic_averages = np.array(topic_averages)
    
    # Cosine similarity
    cs = cosine_similarity(X_test, topic_averages)
    df_hotel_test['closest_topic'] = cs.argmax(axis=1)+1 # Add one because of the 0 indexing
    df_hotel_test['closest_correct'] = df_hotel_test.closest_topic == df_hotel_test.topic_hotel
    scores.append(df_hotel_test.closest_correct.mean())
    
print(scores)
print(np.mean(scores))
```
